```python
#so we have one missing values
# check row which contains null values (Hint axis=1)
rows_with_null = df[df.isnull().any(axis=1)]

# Display the rows with null values
print("Rows with null values:")
print(rows_with_null)
```

```python
#fill null value with most suitable value (hint to replace,use inplace=True)
 #replacing Nan with Montenegro
df.fillna('Montenegro', inplace=True)
```

```python
# Verify that the missing value has been replaced
print("Check if null values are replaced:")
print(df.isnull().sum())
```

```python
#sanity check weather null values are replaced or not
 #checking weather it is replaced or not
print("Sample of the Other names column:")
print(df['Other names'].sample(20))
```

```python
#plot bar graph for top 5 countries which have more number of covid  Death percentage 
#plot bar graph for top 5 countries which have more number of covid Deaths
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(most_countries_deathp['Country'],most_countries_deathp['Death percentage'])
plt.xlabel('Country')
plt.ylabel('Death percentage')
plt.title('Most 5 Countries by COVID death percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Continent-wise cases
continent_cases = df.groupby('Continent')['Total Cases'].sum().reset_index()
print("Continent-wise cases:")
continent_cases
```

```python
#draw bar plot for continent wise case
# Draw bar plot for continent-wise cases
plt.figure(figsize=(10, 6))
plt.bar(continent_cases['Continent'], continent_cases['Total Cases'])
plt.xlabel('Continent')
plt.ylabel('Cases')
plt.title('Continent-wise COVID Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#find relationship between Population and total cases.Comment  on outliers
# Relationship between Population and total cases
plt.figure(figsize=(10, 6))
plt.scatter(df['Population'], df['Total Cases'])
plt.xlabel('Population')
plt.ylabel('Cases')
plt.title('Relationship between Population and Total Cases')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```

```python
#find relationship between Population and total deathss.Comment  on outliers
# Relationship between Population and total cases
plt.figure(figsize=(10, 6))
plt.scatter(df['Population'], df['Total Deaths'])
plt.xlabel('Population')
plt.ylabel('Total Deaths')
plt.title('Relationship between Population and Total death')
plt.tight_layout()
plt.show()
```

```python
# Spearman rank correlation heatmap.Comment on co-relation.
# Spearman rank correlation heatmap
import seaborn as sns
correlation_matrix = df.corr(method='spearman') #calculates the Spearman rank correlation coefficients for all pairs of numerical columns in the DataFrame
# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=.5)
plt.title('Spearman Rank Correlation Heatmap - COVID-19 Dataset')
plt.show()
```
